Import Python Libraries

```
import pandas as pd
import datetime
from pandas import DataFrame
from datetime import timedelta
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

Adjustment for Timestamp (Accounting for Difference of Epoch Time)

```
aa = datetime(1900, 1, 1)
bb = datetime(1970, 1, 1)
cc = timedelta(hours=8)
tdelta = bb-aa-cc
```

# Winter Data Analysis

###Import Data Set for Winter:
### Coastal Endurance â€º Oregon Shelf Surface Piercing Profiler Mooring


```
url = 'https://raw.githubusercontent.com/Branth1/CTD-Project1/master/Dataset1.csv'
df1 = pd.read_csv(url)
df1
```

Define Variables for Plotting

```
x=df1.temperature
y= df1.pressure
S= df1.salinity
z= df1.pressure * 1.019977
x2=x*x
x3=x*x*x
```

Define SSP & Plot

```
c = 1449.2+(4.6*x) - (0.055*x2) + (0.00029*x3) + (1.34-0.01*x)*(S-35)+0.016*z
avg = sum(c)/len(c)
plt.scatter(c,z)
plt.ylabel('Depth (m)')
plt.xlabel('SSP (m/s)')
plt.axvline(x=avg, color='r', linestyle='-')
plt.gca().invert_yaxis()
plt.legend (("Average(1491)","Data"))
plt.title('Coastal Endurance â€º Oregon Shelf Surface Piercing Profiler Mooring 11/01-11/02/18')
print("The average is ", round(avg,2),"m/s")
```

Plot Depth Vs Time to Determine # of Dives

```
df1.time = pd.to_datetime(df1.time,unit='s')
time=df1.time-tdelta
plt.plot(time,z)
degrees = 50
plt.xticks(rotation=degrees)
plt.gca().invert_yaxis()
plt.xlabel('Time')
plt.ylabel('Depth (m)')
plt.title('Time Vs Depth (# of Dives)')
```
