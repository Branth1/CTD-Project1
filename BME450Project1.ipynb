# Winter Data Analysis

###Import Data Set for Winter:
### Coastal Endurance › Oregon Shelf Surface Piercing Profiler Mooring


```
url = 'https://raw.githubusercontent.com/Branth1/CTD-Project1/master/Dataset1.csv'
import pandas as pd
df1 = pd.read_csv(url)
df1
```

Define Variables for Plotting

```
x=df1.temperature
y= df1.pressure
S= df1.salinity
z= df1.pressure * 1.019977
x2=x*x
x3=x*x*x
```

Define SSP & Plot

```
c = 1449.2+(4.6*x) - (0.055*x2) + (0.00029*x3) + (1.34-0.01*x)*(S-35)+0.016*z
```

```
avg = sum(c)/len(c)
print("The average is ", round(avg,2),"m/s")
```

```
import matplotlib.pyplot as plt
plt.scatter(z,c)
plt.xlabel('Depth (m)')
plt.ylabel('SSP (m/s)')
plt.axhline(y=avg, color='r', linestyle='-')
plt.legend (("Average(1491)","Data"))
plt.title('Coastal Endurance › Oregon Shelf Surface Piercing Profiler Mooring 11/01-11/02/18')
```
